```python
import openpyxl
```

```python
workbook = openpyxl.load_workbook('./Crops.xlsx')
```

```python
worksheet = workbook.active
```

```python
for row in worksheet.iter_rows():
    for cell in row:
        if cell.hyperlink:
           
            address = cell.hyperlink.target

           
            cell.value = address
            cell.hyperlink = None 
```

```python
workbook.save("modified_excel_file.xlsx")
```
